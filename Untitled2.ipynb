```python
# Jacobi
matrix = []
tebak = [[]]
batas = float(input('Masukan nilai batas: '))
ralat = batas+1
k = 0
print('\nHey, input matrix yang dah dominal diagonal\n ')
try:
    banyak_variabel_bebas = int(input('jumlah variabel bebas: '))
    assert banyak_variabel_bebas != 0
    for z in range(banyak_variabel_bebas):
        print('\n')
        matrixsementara = [];matrix.append(matrixsementara)
        for r in range(banyak_variabel_bebas + 1):
            try:
                masukan = input(f'a{z + 1}{r + 1}:');matrixsementara.append(float(masukan))
            except ValueError:
                exit('input harus berupa bilangan')
        input_tebakan = input(f"\nnilai tebakan x{z + 1}: ")
        tebak[0].append(float(input_tebakan))

except AssertionError:
    exit('Jumlah variabel bebas tak boleh 0 dan harus bilangan bulat')
except ValueError:
    exit('Jumlah variabel bebas harus bilangan bulat')

print(tebak)

while ralat > batas:
    print(f'\niterasi ke-{k+1}\n')
    jumlah_residu = 0
    tebak_iterasi = []
    tebak.append(tebak_iterasi)
    for i in range(len(matrix)):
        jumlah = 0
        for j in range(len(matrix)):
            jumlah += matrix[i][j]*tebak[k][j]
        residu = (1/matrix[i][i] * (matrix[i][-1]-jumlah))
        jumlah_residu += residu
        lengkap = tebak[k][i] + residu
        print(f'x{i+1}= {lengkap}')
        tebak_iterasi.append(lengkap)
    ralat = (jumlah_residu**2)**0.5
    jumlah_residu = 0
    print(f'ralat = {ralat}')
    k += 1

tebak.clear()

```
